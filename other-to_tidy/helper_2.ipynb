# Journal Impact Factor

```python
import glob
from rapidfuzz import process
import numpy as np
```

```python
# Create df of uniques journals to see how many can be matched
os.chdir('..')
os.chdir(os.getcwd() + '\\consolidate' '\\scopus')
```

```python
dfs = []

for csv in glob.glob('*.csv'):
    dfs.append(pd.read_csv(csv))
```

```python
scopus = pd.concat(dfs)
```

```python
journals = scopus['publicationName'].unique()
```

```python
journals.shape
```

```python
journals = pd.DataFrame(journals, columns=['journal'])
```

```python
# Load JIF data
#os.chdir('..')
jif = pd.read_csv('JCR_2020.csv').drop(columns=['Unnamed: 6', 'Unnamed: 7'])
```

```python
jif.tail()
```

```python
# Match journals against JIF data
jif_journal_names = jif['Full Journal Title'].to_list()
```

```python
% timeit process.extract('assistive tech', jif_journal_names)
```

```python
0.137 * journals.shape[0] / 60
```

```python
journals['fuzzy_match'] = journals['journal'].apply(lambda x: process.extract(x, jif_journal_names))
```

```python
def find_best(x, score=False):
    try:
        if score:
            return x[0][1]
        else:
            return x[0][0]
    except:
        return np.nan
```

```python
journals['best_match'] = journals['fuzzy_match'].apply(find_best)
journals['best_score'] = journals['fuzzy_match'].apply(lambda x: find_best(x, score=True))
```

```python
print(f'There are {journals.shape[0]} journals from Scopus')
print('{} of them have exact matches from JCR data'.format(journals[journals['best_score'] == 100].shape[0]))
```

```python
journals[journals['best_score'] >= 90].shape[0]
```

```python
journals[(journals['best_score'] >= 90) & (journals['best_score'] <= 92)]
```
